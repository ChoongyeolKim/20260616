```python
import datetime

now = datetime.datetime.now()

print(now.year) # 연도
print(now.month) # 월 (1-12)
print(now.day) # 일 (1-31)
print(now.hour) # 시 (0-23)
print(now.minute) # 분 (0-59)
print(now.second) # 초 (0-59)
print(now.microsecond) # 마이크로초 (0-999999)
print(now.tzinfo) # 타임존
```

```python
type(df_monthly.index)
```

```python
print(type(df_monthly.index), df_monthly.index.dtype, df_monthly.index.freq)
```

```python
# NOTE - 다음 코드는 에러남
df_monthly.plot(kind='bar', title='월 평균 대여량')

xtic = df_monthly.index.strftime('%Y-%m')
plt.xticks(range(len(df_monthly.index)), xtic)
```

```python
# 방법 1 - Pandas 그래프 그리기 에러를 우회하기 위해 다음처럼 수동으로 그림
copy1 = df_monthly.copy()

labels = copy1.index.strftime('%Y-%m')
fig, ax = plt.subplots(figsize=(10, 4))

ax.bar(range(len(copy1)), copy1['count'])
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=90)
ax.set_title('월 평균 대여량')

plt.tight_layout()
```

```python
# 방법 2 - Pandas 그래프 그리기 에러를 우회하기 위해 다음처럼 수동으로 그림
copy2 = df_monthly.copy()
copy2.index = copy2.index.strftime('%Y-%m')
copy2.plot(kind='bar', title='월 평균 대여량')

plt.xticks(rotation=90)
plt.tight_layout()
```
