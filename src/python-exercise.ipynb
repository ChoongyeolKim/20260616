## 데이터형 변환 및 출력 포맷팅

```python
# 문자열로 입력 받은 값들을 정수형으로 변환
a = int(input('첫 번째 정수 입력: '))
b = int(input('두 번째 정수 입력: '))

print('{0} * {1} = {2:,}'.format(a, b, a*b))
```

```python
# 문자열로 입력 받은 값들을 실수형으로 변환
c = float(input('첫 번째 실수 입력: '))
d = float(input('두 번째 실수 입력: '))
g = c*d

print(f'{c:.2f} * {d:.3f} = {g:,.2f}')
```

```python
# g 변수에 truncate 된 형대로 저장되어 있는 것이 아니라
# 화면에 표시할 때만 자리수를 지정해서 출력함
print(f'{g:,.5f}')
```

## 출생년도 끝자리와 나이를 입력받아 가능한 요일을 출력

```python
last_number_in_year = int(input('출생년도 끝자리 : '))
age = int(input('나이 : '))

# place holder 를 완성해 보세요.
if age >= 65:
    print('모든 요일')
else:
    available_number = last_number_in_year % 5
    if available_number == 1:
        print('월')
    elif available_number == 2:
        print('화')
    elif available_number == 3:
        print('수')
    elif available_number == 4:
        print('목')
    else:
        print('금')
```

## 날짜에 따른 차량 운행 가능 표시

```python
today = int(input('오늘 날짜 입력: '))

# place holder 를 완성해 보세요.
if 1 <= today <= 31:
    rego_str = input('차량번호 입력: ')
    if len(rego_str) != 4:
        print('차량번호 오류')
    else:
        rego = int(rego_str)
        if rego % 2 == 0: # 짝수
            print('짝수 차량 운행날')
            if today % 2 == 0:
                # 오늘 날짜가 짝수이면
                print('운행 가능')
            else:
                # 오늘 날짜가 홀수이면
                print('운행 불가능')
        else: # 홀수
            print('홀수 차량 운행날')
            if today % 2 == 0:
                # 오늘 날짜가 짝수이면
                print('운행 불가능')
            else:
                # 오늘 날짜가 홀수이면
                print('운행 가능')    
else:
    print('날짜 입력 오류')
```

## 키와 몸무게 입력 후 BMI 지수와 그에 따른 체중 상태 출력

```python
height = float(input('키 입력(cm): '))
weight = int(input('몸무게 입력(kg): '))

# place holder 를 완성해 보세요.
bmi = weight / ((height ** 2) / 10000)
if bmi >= 23:
    print(f'BMI 지수는 {bmi:.2f}로 과체중')
elif bmi >= 18.5:
    print(f'BMI 지수는 {bmi:.2f}로 정상체중')
else:
    print(f'BMI 지수는 {bmi:.2f}로 저체중')
```

## while 을 사용하여 1부터 입력받은 값까지 더하기

```python
endValue = int(input('합을 원하는 정수값 입력 : '))

sum1 = 0
value = 1

# place holder 를 완성해 보세요.
while value <= endValue:
    sum1 += value
    value += 1

print(f'1부터 {endValue}까지의 합은 {sum1:,} 입니다.')
```

## while 을 사용하여 반복문 실행하기

```python
dan = int(input('출력할 단 입력 : '))

# place holder 를 완성해 보세요.
print(f'*** {dan} 단 ***')
index = 1
while index < 10:
    print(f'{dan} * {index} = {dan * index}')
    index += 1
```

```python
sum3 = 0
for i in range(1, 11):
    sum3 = sum3 + i
    if sum3 >= 20:
        print(f'정수: {i}, 합: {sum3}')
        break
```

```python
sum3 = 0
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i, end=' ')
    sum3 = sum3 + i

print(f'\n1~10까지의 정수 중 홀수의 합 : {sum3}')
```
