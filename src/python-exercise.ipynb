## 날짜에 따른 차량 운행 가능 표시

```python
today = int(input('오늘 날짜 입력: '))

# place holder 를 완성해 보세요.
if 1 <= today <= 31:
    rego_str = input('차량번호 입력: ')
    if len(rego_str) != 4:
        print('차량번호 오류')
    else:
        rego = int(rego_str)
        if rego % 2 == 0: # 짝수
            print('짝수 차량 운행날')
            if today % 2 == 0:
                # 오늘 날짜가 짝수이면
                print('운행 가능')
            else:
                # 오늘 날짜가 홀수이면
                print('운행 불가능')
        else: # 홀수
            print('홀수 차량 운행날')
            if today % 2 == 0:
                # 오늘 날짜가 짝수이면
                print('운행 불가능')
            else:
                # 오늘 날짜가 홀수이면
                print('운행 가능')    
else:
    print('날짜 입력 오류')
```

```python
students = [
    {'name':'Alice', 'grades':[90, 85, 92]},
    {'name':'Bob', 'grades':[78, 80, 88]},
    {'name':'Charlie', 'grades':[95, 92, 89]}
]
students
```

```python
average_grade = 0
students_with_higher_grades = []

for student in students:
    student_name = student['name']
    student_grades = student['grades']

    for grade in student_grades:
        if grade >= 90:
            students_with_higher_grades.append(student_name)
            break

    student_average = sum(student_grades)/len(student_grades)
    average_grade += student_average
    print(f"{student_name}'s average grade: {student_average}")

print(f'Average grade of all students: {average_grade/len(students)}')
print(f'Students with grades 90 or higher: {students_with_higher_grades}')
```

```python
def isLeapYear(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

print(f'2022년은 윤년일까? {isLeapYear(2022)}')
print(f'2010년은 윤년일까? {isLeapYear(2010)}')
print(f'2000년은 윤년일까? {isLeapYear(2000)}')
print(f'1900년은 윤년일까? {isLeapYear(1900)}')
```

```python
def add_many(*args):
    total = 0
    for arg in args:
        total += arg
    return total
    # return sum(args)

print(add_many(1, 2))
print(add_many(1, 2, 3))
print(add_many(1, 2, 3, 4, 5, 6, 7, 8, 9))
```

```python
def add_mul(op, *args):
    if op == 'add':
        total = 0
        for arg in args:
            total += arg
    elif op == 'mul':
        total = 1
        for arg in args:
            total *= arg
    else:
        raise ArithmeticError
    
    return total

print(add_mul('add', 1, 2, 3, 4, 5))
print(add_mul('mul', 1, 2, 3, 4, 5))
```

```python
import random

sel = ['가위', '바위', '보']

def checkWin(user, com):
    if user == '가위':
        if com == '보':
            print('이겼음')
            return True
        elif com == '가위':
            print('비겼음')
            return False
        else:
            print('졌음')
            return False
    elif user == '바위':
        if com == '가위':
            print('이겼음')
            return True
        elif com == '바위':
            print('비겼음')
            return False
        else:
            print('졌음')
            return False
    elif user == '보':
        if com == '바위':
            print('이겼음')
            return True
        elif com == '보':
            print('비겼음')
            return False
        else:
            print('졌음')
            return False

    return True

def continueConfirm():
    print('계속 하시겠습니까? (Y/N)')
    answer = input('계속 하시겠습니까? (Y/N)')
    if answer[0].lower() == 'y': # Y,y
        return True
    else:
        return False

while True:
    user = input('가위, 바위, 보: ')
    com = sel[random.randint(0, 2)]
    if checkWin(user, com):
        if continueConfirm():
            continue
        else:
            print('종료합니다.')
            break
    else:
        print('다시 시작합니다.')
```

## Class

```python
class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second
    # def setdata(self, first, second):
    #     self.first = first
    #     self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
    def div(self):
        if self.second == 0:
            return 0
        
        result = self.first / self.second
        return result
```

```python
a = MoreFourCal(4, 2)
print(a.add())
print(a.pow())
```

```python
b = MoreFourCal(3, 0)
print(b.add())
```

```python
print(b.div())
```

```python
c = MoreFourCal(3, 0)
print(c.add())
print(c.div())
```

## JSON

```python
import json

# 파이썬 딕셔너리 데이터
data = {
    'name': '홍길동',
    'age': 30,
    'city': 'Seoul'
}

# 딕셔너리를 JSON 문자열로 변환
print(json.dumps(data, ensure_ascii=True))
print(json.dumps(data, ensure_ascii=False)) # ensure_ascii=False 옵션을 사용하여 한글이 깨져 보이지 않도록 함

json_string = json.dumps(data, ensure_ascii=False)
print(f'JSON 문자열 : {json_string}')
print(type(json_string))

# JSON 문자열을 다시 파이썬 딕셔너리 데이터 타입으로 변환
parsed_data = json.loads(json_string)
print(f'파이썬 데이터 타입 : {type(parsed_data)}')
print(f'파이썬 데이터 : {parsed_data}')
```

## JSON 파일 입출력

```python
# 파일 쓰기
with open('../data/example1.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
# 파일 읽기
with open('../data/example1.json', 'r', encoding='utf-8') as f:
    content = json.load(f)
    print(content)
    print(type(content))
    print('나이:', content['age'])
```

## CSV

```python
import csv

csv_data = [
    ['name', 'age', 'city'],
    ['홍길동', 30, 'Seoul'],
    ['김철수', 25, 'LA'],
    ['이영희', 28, 'Tokyo']
]

# 파일 쓰기
with open('../data/example2.csv', 'w', newline='', encoding='utf-8-sig') as f: # UTF-8 BOM
    writer = csv.writer(f)
    writer.writerows(csv_data) # csv_data 리스트의 모든 행을 한번에 디스크에 씀.
    # 한줄씩 쓰려면 writer.writerow(row) 를 사용한다.
```

```python
# 파일 읽기 - 한 줄씩 읽기
with open('../data/example2.csv', 'r', encoding='utf-8-sig') as f:
    reader = csv.reader(f)
    # next(reader) # 헤더를 건너 뛰려면 이 줄을 추가
    for row in reader:
        print(row)
```

```python
# 파일 읽기 - 전체 읽기
with open('../data/example2.csv', 'r', encoding='utf-8-sig') as f:
    content = f.read()
    print(content)
```

```python
# 인코딩 변환
with open('../data/example2-without-bom.csv', 'w', newline='', encoding='utf-8') as f: # UTF-8
    f.write(content)
```

## Regular Expression (정규표현식)

```python
import re

text = '문의사항은 010-1234-5678 또는 support@example.com 으로 연락주세요. 010-5678-1234'

# 전화번호 패턴 추출
mobile_pattern = r'\d{3}-\d{4}-\d{4}' # \d{3,4} 는 숫자 3자리 또는 4자리를 의미
mobile_numbers = re.findall(mobile_pattern, text)
print('추출된 전화번호: ', mobile_numbers)

# 이메일 패턴 추출
email_pattern = r'[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+'
email_address = re.findall(email_pattern, text)
print('추출된 이메일: ', email_address)
```
