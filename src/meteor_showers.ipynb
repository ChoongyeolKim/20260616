```python
# 달의 위상 데이터 프레임에서 percentage 컬럼의 결측치 처리하기
# 데이터 프레임의 각 행을 순회하면서 이전 위상값으로 채워넣기

# iterrows()를 사용하여 moon_phases 데이터프레임의 각 행을 순차적으로 반복하면서 행(row)과 그 행의 인덱스(index)를 가져온다.
# 결측값인 경우, lastphase에 저장된 값을 해당 행의 'percentage'에 채워 넣는다.
# 즉, 이전에 확인된 마지막 'percentage' 값을 사용하여 결측값을 대체한다.

# 첫번째 행은 데이터가 존재하지 않은데 csv 파일을 열어보면 두번째 행이 'first quarter' 이므로
# 첫번째 행의 percentage 컬럼값은 new moon 에 해당하는 0 이다.

test3 = moon_phases_dropped.copy()

found_first_value = False
for i, row in test3.iterrows():
    if pd.isna(row['percentage']):
        test3.loc[i, 'percentage'] = test3.loc[i-1, 'percentage'] if i > 0 else 0.0

test3
```

```python
# 다음과 같이 처리할 수도 있음
test4 = moon_phases_dropped.copy()

last = 0.0

for i, row in test4.iterrows():
    if pd.isna(row['percentage']):
        test4.loc[i, 'percentage'] = last
    else:
        last = row['percentage']

test4
```

```python
# test3과 test4가 동일한지 확인하기
try:
    pd.testing.assert_frame_equal(test3, test4) # check_like=True
    print("Equal")
except AssertionError as e:
    print("Not equal:", e)
```

```python
# 벡터화 방법 (Vectorization) - ffill() 을 사용하는 방법
test5 = moon_phases_dropped.copy()
test5['percentage'].ffill()
```

```python
def predict_constellations_viewable(city):
    constellations_list = None
    latitude = get_latitude(city)

    if latitude:
        condition1 = latitude >= constellations_dropped['latitudeend']
        condition2 = latitude <= constellations_dropped['latitudestart']
        viewable = (condition1) & (condition2)
        constellations_viewable = constellations_dropped[viewable]

        if not constellations_viewable.empty:
            constellations_list = constellations_viewable['constellation'].to_list()
            return latitude, constellations_list
    
    return latitude, None
```

```python
print(predict_constellations_viewable('Abu Dhabi'))
print(predict_constellations_viewable('Seoul'))
print(predict_constellations_viewable('Jeju'))
print(predict_constellations_viewable('Wellington'))
```

```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기
def predict_best_meteor_showers_viewing(city):
    # 입력한 도시 정보가 존재하지 않으면 오류 메시지
    # 방법 1
    # if city not in cities['city'].values:
    #     return f'입력하신 도시인 {city} 의 정보를 찾을 수 없습니다.'
    
    # 도시의 위도 정보를 불러오기
    # 방법 2
    # latitude = get_latitude(city)
    # if not latitude:
    #     return f'입력하신 도시인 {city} 의 정보를 찾을 수 없습니다.'

    # 방법 3
    latitude, constellations_list = predict_constellations_viewable(city)
    if not latitude:
        # print(f'입력하신 도시인 {city} 의 정보를 찾을 수 없습니다.')
        return f'입력하신 도시인 {city} 의 정보를 찾을 수 없습니다.'

    print(f'입력하신 도시인 {city} 의 위도는 {latitude} 입니다.')

    # 해당 도시의 위도에서 관측 가능한 별자리 조회

    # 볼 수 있는 별자리가 존재하지 않으면 오류 메시지
    if not constellations_list:
        return f'입력하신 도시인 {city} 에서 관찰 가능한 별자리가 없습니다.'
    
    # 별자리 관측 가능 안내 메시지
    meteor_showers_string = f'관찰 가능한 별자리는 다음과 같습니다.\n'
    
    # 도시별 관찰 가능한 별자리 iterate
    for constellation in constellations_list:
        # 별자리와 가장 가까운 유성우 조회
        meteor_showers_viewable = meteor_showers_dropped.loc[meteor_showers_dropped['radiant'] == constellation]
        constellation_name = meteor_showers_viewable.iloc[0]['name'] # 별자리 이름
        
        # 유성우 관측이 가능한 시작일과 종료일
        startdate = meteor_showers_viewable.iloc[0]['startdate']
        enddate = meteor_showers_viewable.iloc[0]['enddate']

        # 유성우가 보일 때의 달의 위상 조회
        condition1 = moon_phases_filled['date'] >= startdate
        condition2 = moon_phases_filled['date'] <= enddate
        condition = condition1 & condition2
        moon_phases_meteor_showers = moon_phases_filled.loc[condition]

        # 달이 눈에 보이지 않는 첫째날 조회
        idx = moon_phases_meteor_showers['percentage'].idxmin()
        best_date = moon_phases_meteor_showers.loc[idx, 'date']
        best_date_str = best_date.strftime('%B %d, %Y')
        
        # 사용자에게 정보 전달
        # Lyrids 를 잘 보려면 Lyra 자리 위치를 향하여 April 22, 2020 에 관찰하면 됩니다.
        meteor_showers_string += f'{constellation_name} 를 잘 보려면 {constellation} 자리 위치를 향하여 {best_date_str} 에 관찰하면 됩니다.\n'

    return meteor_showers_string.rstrip()
```

```python
def predict_constellations_viewable2(df, city):
    latitude = get_latitude(city)

    if latitude:
        condition1 = latitude >= df['latitudeend']
        condition2 = latitude <= df['latitudestart']
        viewable = (condition1) & (condition2)
        constellations_viewable = df[viewable]

        if not constellations_viewable.empty:
            constellations_list = constellations_viewable['constellation'].to_list()
            return latitude, constellations_list
    
    return latitude, None
```
