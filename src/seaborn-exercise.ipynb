```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') # 'Apple Gothic'
plt.rc('axes', unicode_minus=False)
```

```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
sns.barplot(data=tips, x='day', y='tip', estimator='mean', errorbar=None)
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', estimator='sum', errorbar=None)
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', estimator='sum', errorbar=None,
            hue='sex')
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', estimator='sum', errorbar=None,
            hue='time', palette='Set1')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size')
plt.show()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',
            hue='smoker', size='size', col='time', style='smoker')
plt.show()
```

```python
sns.countplot(data=tips, x='size')
plt.show()
```

```python
sns.boxplot(data=tips, x='size', y='tip')
plt.show()
```

```python
sns.boxplot(data=tips, x='size', y='tip',
            hue='time', palette='Set3')
plt.show()
```

```python
sns.violinplot(data=tips, x='size', y='tip')
plt.show()
```

```python
sns.violinplot(data=tips, x='size', y='tip', hue='time')
plt.show()
```

```python
sns.violinplot(data=tips, x='size', y='tip', hue='time', palette='Set2')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=20)
plt.show()
```
