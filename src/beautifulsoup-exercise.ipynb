# 네이버 파이낸스 뉴스

```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://finance.naver.com/news/mainnews.naver'
response = requests.get(base_url)
status_code = response.status_code
if status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    articles = soup.select('.block1')

    print(status_code, len(articles))
else:
    print('요청 실패. 상태 코드:', status_code)
```

```python
for article in articles:
    print(article)
```

```python
article_list = []

for article in articles:
    title = article.select_one('.articleSubject > a').text.strip()
    link = article.select_one('.articleSubject > a').attrs['href']
    content = article.select_one('.articleSummary').contents[0].strip()
    press = article.select_one('.press').text.strip()
    date = article.select_one('.wdate').text.strip()
    
    article_list.append([title, link, content, press, date])

print(len(article_list))
```

# 네이버 파이낸스 뉴스 URL

- https://finance.naver.com/news/mainnews.naver
- https://finance.naver.com/news/mainnews.naver?&page=2
- https://finance.naver.com/news/mainnews.naver?&page=3

```python
import time

base_url = 'https://finance.naver.com/news/mainnews.naver'
article_list = []

for page_num in range(1, 5):
    news_url = f'{base_url}?&page={page_num}'
    response = requests.get(news_url)
    status_code = response.status_code
    if status_code == 200:
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
        articles = soup.select('.block1')

        for article in articles:
            title = article.select_one('.articleSubject > a').text.strip()
            link = article.select_one('.articleSubject > a').attrs['href']
            content = article.select_one('.articleSummary').contents[0].strip()
            press = article.select_one('.press').text.strip()
            date = article.select_one('.wdate').text.strip()
            
            article_list.append([title, link, content, press, date])

        print(f'{page_num}번째 페이지 - 기사 개수 {len(articles)}')
    else:
        print('요청 실패. 상태 코드:', status_code)
    
    time.sleep(5)

print(f'전체 기사 개수 {len(article_list)}')
```

# 마지막 페이지 체크 코드

```python
import time

base_url = 'https://finance.naver.com/news/mainnews.naver'
article_list = []

page_num = 1
while True:
    news_url = f'{base_url}?&page={page_num}'
    response = requests.get(news_url)
    status_code = response.status_code
    if status_code == 200:
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
        articles = soup.select('.block1')

        for article in articles:
            title = article.select_one('.articleSubject > a').text.strip()
            link = article.select_one('.articleSubject > a').attrs['href']
            content = article.select_one('.articleSummary').contents[0].strip()
            press = article.select_one('.press').text.strip()
            date = article.select_one('.wdate').text.strip()
            
            article_list.append([title, link, content, press, date])

        print(f'{page_num}번째 페이지 - 기사 개수 {len(articles)}')

        # pgRR
        last_page_marker = soup.select_one('.pgRR')
        if last_page_marker is None:
            break
    else:
        print('요청 실패. 상태 코드:', status_code)
        break
    
    page_num += 1
    time.sleep(5)

print(f'전체 기사 개수 {len(article_list)}')
```
