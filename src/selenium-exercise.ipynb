# 전체 코드

```python
import time
from selenium.webdriver.common.by import By

base_url = 'https://finance.naver.com/news/mainnews.naver'
article_list = []

for page_num in range(1, 7):
    news_url = f'{base_url}?&page={page_num}'
    driver.get(news_url)

    elements = driver.find_elements(By.CSS_SELECTOR, '.block1')
    for element in elements:
        title = element.find_element(By.CSS_SELECTOR, '.articleSubject > a').text.strip()
        link = element.find_element(By.CSS_SELECTOR, '.articleSubject > a').get_attribute('href')
        tokens = element.find_element(By.CSS_SELECTOR, '.articleSummary').text.split()[:-4]
        content = str.join(' ', tokens)
        press = element.find_element(By.CSS_SELECTOR, '.press').text.strip()
        date = element.find_element(By.CSS_SELECTOR, '.wdate').text.strip()

        article_list.append([title, link, content, press, date])

    print(f'{page_num}번째 페이지 - 기사 개수 {len(elements)}')

    time.sleep(5)

print(f'전체 기사 개수 {len(article_list)}')
```
