```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') # 'Apple Gothic'
plt.rc('axes', unicode_minus=False)
```

```python
age = [7, 8, 9, 10, 11]
height = [122.1, 127.9, 133.4, 138.8, 142.7]

plt.plot(age, height, color='red', marker='o', ls=':')
plt.grid(ls='--', color='skyblue')
plt.title('나이에 따른 키')
plt.xlabel('나이')
plt.ylabel('키')

plt.show()
```

```python
blood_type=['A', 'B', 'O', 'AB']
count=[45, 35, 15, 5]

plt.bar(blood_type, count, color=['red', 'blue', 'green', 'purple'])
plt.show()
```

```python
plt.barh(blood_type, count, color=['red', 'blue', 'green', 'purple'])
plt.show()
```

```python
plt.pie(count, labels=blood_type, autopct='%.1f%%', colors=['red', 'blue', 'green', 'purple'])
plt.show()
```

```python
import seaborn as sns
```

```python
seaborn_data_list = sns.get_dataset_names()
print(seaborn_data_list)
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
x = tips['total_bill']
y = tips['tip']

plt.scatter(x, y)
plt.xlabel('total_bill')
plt.ylabel('tip')

plt.show()
```

```python
plt.hist(tips['total_bill'])
plt.show()
```

```python
plt.hist(tips['total_bill'], bins=20)
plt.show()
```

```python
plt.boxplot(tips['total_bill'])
plt.show()
```

```python
import pandas as pd

df1 = pd.DataFrame({'월':[1, 2, 3, 4, 5], '몸무게':[80, 78, 75, 73, 70]})
df2 = pd.DataFrame({'월':[1, 2, 3, 4, 5], '몸무게':[60, 62, 59, 55, 54]})
```

```python
plt.plot(df1['월'], df1['몸무게'], label='철수')
plt.plot(df2['월'], df2['몸무게'], label='영희')
plt.legend()

plt.show()
```
